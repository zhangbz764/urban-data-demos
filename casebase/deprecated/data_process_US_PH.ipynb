```python
import fiona

gdb_path = r"E:\2_data\building_3d_opensource\Philadelphia2015_globe.gdb\Philadelphia2015_globe.gdb"

# 查看所有图层
layers = fiona.listlayers(gdb_path)
print("图层列表：")
for layer in layers:
    print(f"  {layer}")
```

```python
with fiona.open(gdb_path, layer="Buildings") as src:
    print(f"坐标系: {src.crs}")
    print(f"几何类型: {src.schema['geometry']}")
    print(f"字段: {src.schema['properties']}")
    print(f"总数: {len(src)}")
    
    # 只看第一条记录的属性，不读几何
    first = next(iter(src))
    print(f"\n第一条属性:")
    for k, v in first['properties'].items():
        print(f"  {k}: {v}")
```
